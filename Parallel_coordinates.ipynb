```python
%reset
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import ticker
#%matplotlib inline
df = pd.read_csv('log4.txt', sep = " ")

df = df.drop(columns=['folder'])
df.sort_values(by='mIoU_aver', inplace=True)
df.reset_index(inplace=True, drop=True)
#df['mIoU_aver'] = pd.to_numeric(df['mIoU_aver'].replace('?', np.nan))
df['mIoU_aver'] = pd.cut(df['mIoU_aver'], [0.0, 10.0, 20.0, 30.0, 40.0, 50.0, 60.0, 70.0, 80.0, 90.0, 100.0])


#df.drop(columns=['folder'])

from matplotlib import ticker

# cols = list(df.columns)
# print(cols)
#df.drop('mIoU_aver', axis='columns', inplace=True) #['displacement', 'cylinders', 'horsepower', 'weight', 'acceleration']

# df.drop('IoU_other_aver', axis='columns', inplace=True)
# df.drop('IoU_ground_aver', axis='columns', inplace=True)
# df.drop('IoU_bui_aver', axis='columns', inplace=True)
# df.drop('mIoU_var', axis='columns', inplace=True)
# df.drop('IoU_other_var', axis='columns', inplace=True)
# df.drop('IoU_ground_var', axis='columns', inplace=True)
# df.drop('IoU_bui_var', axis='columns', inplace=True)
# df.drop('acc_aver', axis='columns', inplace=True)
# df.drop('acc_var', axis='columns', inplace=True)
# df.drop('threads', axis='columns', inplace=True)

# df.drop('fsd', axis='columns', inplace=True)
# df.drop('rep_ext', axis='columns', inplace=True)
# df.drop('kp_ext', axis='columns', inplace=True)
# df.drop('max_epoch', axis='columns', inplace=True)
# df.drop('steps_per_epoch', axis='columns', inplace=True)


cols = list(df.columns)
print(cols)
#cols.remove('index')

cols.remove('mIoU_aver') #['displacement', 'cylinders', 'horsepower', 'weight', 'acceleration']
cols.remove('IoU_other_aver')
cols.remove('IoU_ground_aver')
cols.remove('IoU_bui_aver')
cols.remove('mIoU_var')
cols.remove('IoU_other_var')
cols.remove('IoU_ground_var')
cols.remove('IoU_bui_var')
cols.remove('acc_aver')
cols.remove('acc_var')
cols.remove('threads')
cols.remove('max_epoch')
cols.remove('steps_per_epoch')
print(df)
print(cols)

x = [i for i, _ in enumerate(cols)]

colours_list = ['#EEEEEE', '#E7E7E7', '#DDDDDD', '#D7D7D7', '#999999', '#777777', '#555555', '#333333', '#000000', '#000000' ]
#colours_list = ['#000000', '#111111', '#222222', '#333333', '#444444', '#555555', '#666666', '#777777', '#888888', '#999999' ]
#colours_list = ['#000000', '#AA0000', '#00AA00', '#0000AA', '#00AAAA', '#AA00AA', '#AAAA00', '#A0A0A0', '#0A0A0A', '#AAA000' ]
#colours_list = ['#000000', '#AA0000', '#00AA00', '#0000AA', '#00AAAA', '#AA00AA']
# colours = {}
# for i, _ in enumerate(df['mIoU_aver'].cat.categories):
#     colours[df['mIoU_aver'].cat.categories[i]].append(colours_list[i])
# colours = {}
# for i, _ in enumerate(df['mIoU_aver'].cat.categories):
#     print(i)
#     colours[df['mIoU_aver'].cat.categories[i]] = colours_list[i]
    
# create dict of categories: colours
colours = {df['mIoU_aver'].cat.categories[i]: colours_list[i] for i, _ in enumerate(df['mIoU_aver'].cat.categories)}
colours
# Create (X-1) sublots along x axis
fig, axes = plt.subplots(1, len(x)-1, sharey=False, linewidth=30, figsize=(15,5))

# Get min, max and range for each column
# Normalize the data for each column
min_max_range = {}
for col in cols:
    #print('col', col)
    min_max_range[col] = [df[col].min(), df[col].max(), np.ptp(df[col])]
    #print('min_max_range[col]', min_max_range[col]) 
    df[col] = np.true_divide(df[col] - df[col].min(), np.ptp(df[col]))
    #print('df[col]')
    #print(df[col])

for idx in df.index:
    df.loc[idx, cols] = df.loc[idx, cols] + 0.001 * (idx - int(len(df.index)/2))
    #print(idx, idx - int(len(df.index)/2), df.loc[idx, cols])
    
#print(df)

# Plot each row
for i, ax in enumerate(axes):
    for idx in df.index:
        #print(df.loc[idx, cols[i]], ax, i)
        #df.loc[idx, cols[i]] += 0.05
        mpg_category = df.loc[idx, 'mIoU_aver']
        ax.plot(x, df.loc[idx, cols], colours[mpg_category], linewidth=0.2)
        #print(df.loc[idx, cols], colours[mpg_category])
    ax.set_xlim([x[i], x[i+1]])
    
# Set the tick positions and labels on y axis for each plot
# Tick positions based on normalised data
# Tick labels are based on original data
def set_ticks_for_axis(dim, ax, ticks):
    min_val, max_val, val_range = min_max_range[cols[dim]]
    print('min_val, max_val, val_range', min_val, max_val, val_range)
    step = val_range / float(ticks-1)
    tick_labels = [round(min_val + step * i, 2) for i in range(ticks)]
    norm_min = df[cols[dim]].min()
    norm_range = np.ptp(df[cols[dim]])
    norm_step = norm_range / float(ticks-1)
    print('norm_min, norm_range, norm_step', norm_min, norm_range, norm_step)
    ticks = [round(norm_min + norm_step * i, 2) for i in range(ticks)]
    ax.yaxis.set_ticks(ticks)
    ax.set_yticklabels(tick_labels)

for dim, ax in enumerate(axes):
    ax.xaxis.set_major_locator(ticker.FixedLocator([dim]))
    set_ticks_for_axis(dim, ax, ticks=6)
    ax.set_xticklabels([cols[dim]])
    

# Move the final axis' ticks to the right-hand side
ax = plt.twinx(axes[-1])
dim = len(axes)
ax.xaxis.set_major_locator(ticker.FixedLocator([x[-2], x[-1]]))
set_ticks_for_axis(dim, ax, ticks=6)
ax.set_xticklabels([cols[-2], cols[-1]])


# Remove space between subplots
plt.subplots_adjust(wspace=0)

# Add legend to plot
plt.legend(
    [plt.Line2D((0,1),(0,0), color=colours[cat]) for cat in df['mIoU_aver'].cat.categories],
    df['mIoU_aver'].cat.categories,
    bbox_to_anchor=(1.2, 1), loc=2, borderaxespad=0.)

plt.title("Values of mIoU_aver")

plt.show()
fig.savefig('graph.png', dpi=1200)
```

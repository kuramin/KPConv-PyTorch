```python
%reset
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import ticker
#%matplotlib inline
df = pd.read_csv('log_longtrain.txt', sep = " ")
df
```

```python
# How does IoU_other_aver, IoU_ground_aver and IoU_bui_aver change when moving from def_fit_power 0.5 to 1.0
df_fit_pow05 = df[df.index % 2 == 0]
df_fit_pow10 = df[df.index % 2 == 1]
df_fit_pow05.index = df_fit_pow05.index / 2
df_fit_pow10.index = (df_fit_pow10.index - 1) / 2
df_fit_pow_compar = pd.concat([df_fit_pow10['IoU_other_aver'] - df_fit_pow05['IoU_other_aver'],
                     df_fit_pow10['IoU_ground_aver'] - df_fit_pow05['IoU_ground_aver'], 
                     df_fit_pow10['IoU_bui_aver'] - df_fit_pow05['IoU_bui_aver']], axis=1)
df_fit_pow_compar
```

```python
# How does IoU_other_aver, IoU_ground_aver and IoU_bui_aver change when moving num_ker_pts through [15, 25, 35, 45]
# Every row contains comparable values - average IoU of some class, num_ker_pts is different, values of all other params are the same
df_grow_nkpts_IoU_other_aver = np.vstack([np.array(df_fit_pow05['IoU_other_aver'].iloc[0:4]),  np.array(df_fit_pow05['IoU_other_aver'].iloc[4:8]), np.array(df_fit_pow05['IoU_other_aver'].iloc[8:12])])
df_grow_nkpts_IoU_other_aver
```

```python
# How does IoU_other_aver, IoU_ground_aver and IoU_bui_aver change when moving num_ker_pts through [15, 25, 35, 45]
df_grow_nkpts_IoU_ground_aver = np.vstack([np.array(df_fit_pow05['IoU_ground_aver'].iloc[0:4]),  np.array(df_fit_pow05['IoU_ground_aver'].iloc[4:8]), np.array(df_fit_pow05['IoU_ground_aver'].iloc[8:12])])
df_grow_nkpts_IoU_ground_aver
```

```python
# How does IoU_other_aver, IoU_ground_aver and IoU_bui_aver change when moving num_ker_pts through [15, 25, 35, 45]
df_grow_nkpts_IoU_bui_aver = np.vstack([np.array(df_fit_pow05['IoU_bui_aver'].iloc[0:4]),  np.array(df_fit_pow05['IoU_bui_aver'].iloc[4:8]), np.array(df_fit_pow05['IoU_bui_aver'].iloc[8:12])])
df_grow_nkpts_IoU_bui_aver
```
